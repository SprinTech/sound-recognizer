```python
from glob import glob
from tqdm import tqdm
df = pd.DataFrame()

DATA_DIR = '../../data/01/genres/'
genres = os.listdir(DATA_DIR)

# TODO Ajouter un commentaire pour cette partie de code
for genre in genres:
    print(f"[INFO] Start processing {genre} files...")
    files = glob(os.path.join(DATA_DIR, genre, '**'))
    for file in tqdm(files):
        tmp_datas = []
        data, sr = librosa.load(file)
        for offset in range(0, len(data), sr):
            start = offset
            end = offset + sr * 3
            chunk = data[start:end]
            if(len(chunk) == sr * 3):
                tmp_datas.append(chunk)
        for i, y in enumerate(tmp_datas):
            data = extract_features(y, sr)
            filename = f"{os.path.basename(file).strip('.wav')}_{i}.wav"
            data.update({
                "filename": filename,
                "genre": genre,
            })
            df = df.append(data, True)
```

```python
df.to_csv("../../data/02/audio_features.csv")
```
